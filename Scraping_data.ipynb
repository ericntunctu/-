```python
def k(q):
    """
    一次生出一排號碼
    """
    # n text 
    l=[]
    for i in range(1,7):
        # normal number
        nn='//*[@id="Lotto649Control_history_dlQuery_SNo'+str(i)+'_'+str(q)+'"]'
        g=fxtext(nn)
        l.append(g)
    special=fxtext('//*[@id="Lotto649Control_history_dlQuery_No7_'+str(q)+'"]')
    l.append(special)
    
    
    return l
```
